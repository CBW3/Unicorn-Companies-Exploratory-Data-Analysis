```python
#Exploratory Data Analysis of Unicorn Companies File
#importing libraries
import pandas as pd 
import numpy as np 
import datetime
import matplotlib.pyplot as plt 
```

```python
#converting data joined column to datetime
companies["Date Joined"] = pd.to_datetime(companies["Date Joined"])
```

```python
#confirming I did it correctly
companies.info()
```

```python
#attempting to extract year component from date joined column
companies["Year Joined"] = companies["Date Joined"].dt.year
```

```python
#confirming action performed correctly
companies.head()
```

```python
#attempting to take random sample of large dataset 
companies_sample = companies.sample(n = 50, random_state = 42)
```

```python
#visualizing the time it took companies to reach unicorn status 
#creating bar chart
plt.bar(companies_sample["Industry"], companies_sample["Year Joined"] - companies_sample["Year Founded"])
#adding title
plt.title("Bar plot of maximum years taken by company to become unicorn per industry (from sample)")
#setting x-axis label
plt.xlabel("Industry")
#setting y-axis label 
plt.ylabel("Maximum number of years")
#rotating x-axis labels slightly just in case they get in the way of each other and overlap
plt.xticks(rotation=45, horizontalalignment='right')
#displaying the visual created
plt.show()
```

```python
#just took a look at industries, but now I want to explore representative companies within each industry
#creating another bar chart
plt.bar(companies_sample["Industry"], companies_sample["Valuation"])
#title
plt.title("Bar plot of maximum unicorn company valuation per industry (from sample)")
#x-axis
plt.xlabel("Industry")
#y-axis
plt.ylabel("Maximum valuation in billions of dollars")
#rotation of x-axis label
plt.xticks(rotation=45, horizontalalignment='right')
#displaying visual
plt.show()
```

```python
#revisiting this EDA to do more stuff, so importing libraries again just in case
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#dropping duplicates
companies.drop_duplicates().shape
```

```python
#revisiting the data types in the columns
companies.dtypes
```

```python
#sorting data, wanna peek at most recent entries in particular
companies.sort_values(by="Year Founded", ascending=False).head(10)
```

```python
#wanna determine general distribution of unicorn companies founded in each year
companies["Year Founded"].value_counts().sort_values(ascending=False)
```

```python
#since i already created data type format, i wanna explore names of months when companies obtained unicorn status
companies["Month Joined"] = companies["Date Joined"].dt.month_name()
```

```python
#checking if performed correctly
companies.head()
```

```python
#going to create new column that determines how many years it took every company to reach unicorn status
companies["Years To Join"] = companies["Date Joined"].dt.year - companies["Year Founded"]
```

```python
#checking if performed correctly again
companies.head()
```

```python
#going to filter the dataset, particularly interested in year 2021 since it is the most recent in dataset
companies_2021 = companies[companies["Date Joined"].dt.year == 2021]
```

```python
#viewing the data
companies_2021.head()
```

```python
#inserting week joined for 2021 companies, wanna see distribution among weeks now
companies_2021.insert(3, "Week Joined", companies_2021["Date Joined"].dt.strftime('%Y-W%V'), True)
```

```python
#aggregating stuff
companies_by_week_2021 = companies_2021.groupby(by="Week Joined")["Company"].count().reset_index().rename(columns={"Company":"Company Count"})
```

```python
#viewing data
companies_by_week_2021.head()
```

```python
#rather than exploring weeks, i want to continue to compare trends over time but this time by quarter
companies_2020 = companies[companies["Date Joined"].dt.year == 2020]
```

```python
#concatenating subsets 
companies_2020_2021 = pd.concat([companies_2020, companies_2021.drop(columns="Week Joined")])
```

```python
#adding quarter joined column
companies_2020_2021["Quarter Joined"] = companies_2020_2021["Date Joined"].dt.to_period('Q').dt.strftime('%Y-Q%q')
```

```python
#converting valuation column to proper datatypes
companies_2020_2021["Valuation"] =  companies_2020_2021["Valuation"].apply(lambda v: v.strip("$B")).astype(float)
```

```python
#creating new dataframe that i have been trying to seek after this whole time
companies_by_quarter_2020_2021 = companies_2020_2021.groupby(by="Quarter Joined")["Valuation"].mean().reset_index().rename(columns={"Valuation":"Average Valuation"})
```

```python
#finally, viewing the data
companies_by_quarter_2020_2021.head()
```

```python
#now that the setup is complete, i wanna create a series of box plots to visualize distribution of how long 
#it took companies to become unicorns with respect to the month they joined
#creating list that contains months in chronological order
month_order = ["January", "February", "March", "April", "May", "June",
               "July", "August", "September", "October", "November", 
               "December"]
```

```python
#printing month list to make sure its correct
print(month_order)
```

```python
#going to use seaborn to create box plots visualizations
#basic box plot code
sns.boxplot(x=companies['Month Joined'], 
            y=companies['Years To Join'], 
            order=month_order, 
            showfliers=False)
#adding title
plt.title('Distribution of years to become unicorn with respect to month joined')
#rotating labels slightly so no potential overlap on x-axis
plt.xticks(rotation=45, horizontalalignment='right')
#viewing visual
plt.show()
```

```python
#now, i am going to create a bar plot to visualize average number of years it took companies
#to reach unicorn status with respect to when they were founded 
#setting the size of the plot 
plt.figure(figsize=(10,6))
#basic bar graph code
sns.barplot(x=companies["Year Founded"], y=companies["Years To Join"], ci=False)
#adding title
plt.title("Bar plot of years to join with respect to year founded")
#x-axis label
plt.xlabel("Year founded")
#y-axis label
plt.ylabel("Years to join unicorn status")
#rotating x-axis labels so no potential overlap 
plt.xticks(rotation=45, horizontalalignment='right')
#displaying the plot 
plt.show()
```

```python
#im not sure what they future warning is about but code ran regardless so ill just move on
#i wanna revisit those 2021 unicorn companies since they are most recent
#my goal is to create a bar plot to visualize the number of comapnies that joined per interval for 2021
#setting size of plot 
plt.figure(figsize = (20, 5))
#basic bar plot
plt.bar(x=companies_by_week_2021['Week Joined'],height=companies_by_week_2021['Company Count'])
plt.plot()
#x-axis labels 
plt.xlabel("Week number")
#y-axis labels
plt.ylabel("Number of companies")
#creating title
plt.title("Number of companies that became unicorns per week in 2021")
#rotating x-axis labels so no potential overlap
plt.xticks(rotation = 45, horizontalalignment='right', fontsize=8)
#displaying the plot
plt.show()
```

```python
#okay, now i want to create a visual specific to 2021 companies growth over time
#plan is to create group bar chart by quarter that shows company valuation in 2020 as well as 2021 per quarter
#using slicing method to extract year component and time interval 
companies_by_quarter_2020_2021['Quarter Number'] = companies_by_quarter_2020_2021['Quarter Joined'].str[-2:]
companies_by_quarter_2020_2021['Year Joined'] = companies_by_quarter_2020_2021['Quarter Joined'].str[:4]
#setting size of the plot
plt.figure(figsize = (10, 5))
#creating the basic grouped bar chart framework
sns.barplot(x=companies_by_quarter_2020_2021['Quarter Number'],
            y=companies_by_quarter_2020_2021['Average Valuation'],
            hue=companies_by_quarter_2020_2021['Year Joined'])
plt.plot()
#x-axis labels 
plt.xlabel("Quarter number")
#y-axis labels
plt.ylabel("Average valuation (billions of dollars)")
#chart title
plt.title("Average valuation of companies that became unicorns per quarter in 2020 vs. 2021")
#displaying visual
plt.show()
```
